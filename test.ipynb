```python
# functions
def rmse(actual, predicted):
    sum_error = 0.0
    for i in range(len(actual)):
        summ_error += (predicted[i] - actual[i])**2
    mean_error = sum_error/len(actual)
    return sqrt(mean_error)

def eval_algo(algoritm,dataset):
    test_set= []
    for row in dataset:
        row_copy = list(row)
        test_set.append(row_copy)
        row_copy[-1] = None
        test_set.append(row_copy)
    predicted = algoritm(test_set)
    print(predicted)
    actual = [row[-1] for row in dataset]
    rmse_value = rmse(actual, predicted)
    return rmse_value

def mean(values):
    return sum(values) / float(len(values))

def covariance(x,mean_x,y,mean_y):
    covar = 0.0
    for i in range(len(x)):
        covar += (x[i] - mean_x) * (y[i] - mean_y)
    return covar/float(len(x))

def variance(values, mean):
    return (sum((x-mean)**2 for x in values))/float(len(values))

def coeffients(dataset):
    x = [row[0] for row in dataset]
    y = [row[-1] for row in dataset]

    x_mean,y_mean = mean(x),mean(y)
    b1 = covariance(x,x_mean,y,y_mean)/variance(x,x_mean)
    b0 = y_mean - b1*x_mean
    return [b0,b1]

def simple_linear_regression(train,test):
    predictions=[]
    b0,b1 = coeffients(train)
    for row in test:
        y_val = b0+ b1* row [0]
        predictions.append(y_val)
    return predictions

```
